## Student t- distributions

- This distributions look like a normal distribution but with fatter tails(more spread at the ends)
- Family of curves where shape depends on degree of freedom(df=n-1)
- As n tends to infinity, t distribition tends to be normal distribution.
- Uses for small sample less than 30(n<30), unknown population and standard deviations


```python
#creating dataframe
import pandas as pd
data=pd.read_csv('NHANES_BMI_clean_data.csv')#
data.head()
```

### Visualization: Normal bell Cure Vs. fatter tails curve

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, t


# Filter and Sample from Data
filtered = data[(data['BMXLEG'] >= 30) & (data['BMXLEG'] <= 40)]
sample = filtered['BMXLEG'].sample(n=30, replace=False, random_state=42)

#Compute Statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
sample_size = len(sample)
df = sample_size - 1
standard_error = sample_std / np.sqrt(sample_size)

#X-axis Range
x = np.linspace(sample_mean - 4*standard_error, sample_mean + 4*standard_error, 1000)

# PDFs
normal_pdf = norm.pdf(x, loc=sample_mean, scale=standard_error)
t_pdf = t.pdf(x, df=df, loc=sample_mean, scale=standard_error)

#Plot
plt.figure(figsize=(10, 6))
plt.plot(x, normal_pdf, label='Normal Distribution', color='blue')
plt.plot(x, t_pdf, label=f't-Distribution (df={df})', color='orange', linestyle='--')

plt.title('Normal Distribution vs. t-Distribution (Heavier Tails)')
plt.xlabel('BMXLEG (cm)')
plt.ylabel('Density')
plt.axvline(sample_mean, color='green', linestyle='--', label=f'Sample Mean = {sample_mean:.2f}')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

#### Interpretation

- t-distribution has fatter tails, which helps account for extra uncertainty in small samples.

- When n  30, the t-distribution approximates the normal closely.

- This helps you understand why t-tests are preferred over z-tests with smaller samples.


## t-Score

- used when the population standard deviation is unknown, and the sample size is small(n<30) or when you're using
  sample data to estimate the population mean.
- Formula: t-score = (sample mean - population mean)/ (sample standard deviation/ sqrt(sample size))

  

#### t-score to find the location of data

##### locating data points from 30 to 40

```python
import numpy as np
from scipy import stats
#creating sample
#filtered = data[(data['BMXLEG'] >= 30) & (data['BMXLEG'] <= 40)]
#sample_data = filtered['BMXLEG'].sample(n=30, replace=False, random_state=42)

filtered= data[(data['BMXLEG']>=30 ) & (data['BMXLEG'] <=40) ]
#creating sample of size 30
sample_data=filtered['BMXLEG'].sample(n=30,replace=False,random_state=42)
sample_mean=np.mean(sample_data)
population_mean=np.mean(data['BMXLEG'])
sample_std=np.std(sample_data,axis=0,ddof=1) #axis for dataframe
#above std give 'TypeError: unsupported format string passed to Series.__format__'
#sample_std=np.std(sample_data['BMXLEG'],axis=0,ddof=1) #give KeyError: 'BMXLEG' for series

sample_size=len(sample_data)

#t-score
t_score=(sample_mean - population_mean ) / (sample_std/np.sqrt(sample_size))
percentile=stats.norm.sf(abs(t_score))  *100

print(f'sample mean: {sample_mean:.2f}')
print(f'population mean: {population_mean:.2f}')
print(f'sample standard deviation: {sample_std:.2f}' )
print(f'sample size: {sample_size}')
print(f't-score: {t_score:.2f}')
print(f'percentile: {percentile:.2f}')
```

#### Visualizations

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Simulate x-axis
x_values = np.linspace(population_mean - 4*sample_std, population_mean + 4*sample_std, 1000)
y_values = norm.pdf(x_values, population_mean, sample_std / np.sqrt(sample_size))  # SE as std for sampling dist


# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label='Sampling Distribution\n(under H0)', color='blue')

# Vertical lines
plt.axvline(population_mean, color='green', linestyle='--', label=f'Population Mean = {population_mean:.2f}')
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean = {sample_mean:.2f}\nT = {t_score:.2f}')

# Shade area beyond t-score
if sample_mean > population_mean:
    plt.fill_between(x_values, y_values, 0, where=(x_values >= sample_mean), color='red', alpha=0.3, label='p-value region')
else:
    plt.fill_between(x_values, y_values, 0, where=(x_values <= sample_mean), color='red', alpha=0.3, label='p-value region')

# Labels and display
plt.title('T-Score Visualization on Sampling Distribution')
plt.xlabel('BMXLEG (cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

## T-test

A t-test is a statistical test used to determine if there if there is a significant difference between the means of two groups or between a sample mean and a known population mean.Tt's widely used in hypothesis testing when the data is normally disributed, and sample size.

#### Types of t-tests:

1. ###### One-sample t-test:
   - Compares the mean of a single sample to a known value(often the population mean)
2. ###### Independent two-sample t-test:
   - Compare the mean of two independent groups.
   - Assumes that the two groups have a equal variance(Homogeneity of variance) and are indepedent of each other

3. ###### Paired sample t-test:
   - Compares the means from the same group at two different times(e.g. before and after treatment), or from matched pairs
   - Used when the data points in the two groups are dependent or related.

### One-sample t-test

Use One-sample t-test when:
- You have one sample of numeric data.
- The population standard deviation is unknown
- The population is approximately normally distributed

Hypothesis:
let:
 - mu = true population mean
 - mu0 = hypothesized population mean

###### Two-tailed:
  - H0: mu = mu0
  - H1: mu != mu0
###### One-tailed(greater):
  - H0: mu <= mu0
  - H1: mu > mu0
###### One-tailed(less):
  - H0: mu >= mu0
  - H1: mu < mu0

#### Hypothesis:
- Null Hypothesis: mu = 38.13(mu_0)
- Alternate Hypothesis: mu != 38.13(mu_0) (two tails)

```python
from scipy import stats

#sample data
sample=data['BMXLEG'].sample(n=30,replace=False,random_state=42)
mu_0=38.13 #hypothesized mean >> already calculated

n=30
#-----Using Manually---------------------------------#
#sample mean
sample_mean=np.mean(sample)

#sample std dev
sample_std=np.std(sample,ddof=1)

#t-statistics formula
t_stat=(sample_mean-mu_0)/(sample_std/np.sqrt(n))

#df
df=n-1

# two-tailed p-value from t-distribution
p_value= 2 * stats.t.sf(np.abs(t_stat),df)

#-------Using Build in Function -------------------#

t_stat,p_value = stats.ttest_1samp(sample,popmean=mu_0)

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Sample Size: {n}")
print(f"T-statistic: {t_stat:.3f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")
```

#### Interpretation:
- Since p = 0.4440> 0.05, we fail to reject the null hypothesis at the 5% significance level.
- There is no significant evidence that the sample mean differs from the population mean.


```python
import matplotlib.pyplot as plt

df = len(sample) - 1
x = np.linspace(-4, 4, 1000)
t_pdf = stats.t.pdf(x, df)

plt.figure(figsize=(10, 5))
plt.plot(x, t_pdf, label=f't-distribution (df={df})', color='blue')
plt.axvline(t_stat, color='red', linestyle='--', label=f't = {t_stat:.2f}')
plt.fill_between(x, t_pdf, 0, where=(x <= t_stat), color='red', alpha=0.3)
plt.axvline(0, color='black', linestyle='--')
plt.title('One-Sample t-Test: t-distribution and test statistic')
plt.xlabel('t-score')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Independent two Sample t-test

An independent two-sample t-test is used to determine whether the mean of two inpdependet groups are significantly different.
This test assumes:

- Independence: Group have different sujects
- Normality: Each group's data is approximately normally distributed.
- Equal variance(for standard verision)

```python
#creating two different sample groups
# Shuffling the full data
shuffled = data['BMXLEG'].sample(frac=1, random_state=42).reset_index(drop=True)

# Creating two distinct groups
group1 = shuffled[:30]
group2 = shuffled[30:50]  # 20 more, no overlap
```

To determine wether the two groups have equal variance before choosing between student's t-test and welch's t-test, we commonly use levene's test.

```python
#levene test
from scipy.stats import levene
stat,p=levene(group1,group2)

if p>0.05:
    print("Fail to reject H_0: Variances are equal > Use Student's t-test")

else:
    print("Reject H_0: Variance are not equal > Use Welch's t-test")
```

#### Hypothesis
- Null Hypothesis:       mu_1 = mu_2
- Alternate Hypothesis:  mu_1 != mu_2 (two tail test)

```python
###---------calculating manual-----
mean1,mean2=np.mean(group1),np.mean(group2)
std1,std2 =np.std(group1,ddof=1),np.std(group2,ddof=1)
n1,n2= len(group1), len(group2)

#pooled standard deviation
sp=np.sqrt(((n1-1)*std1**2 + (n2-1)*std2**2) / (n1+n2-2))

#t_statistics
t_stat=(mean1-mean2)/(sp*np.sqrt(1/n1+1/n2))

#degree of freedom
df= n1+n2 -2

#two-tailed p-value
p_value=2*stats.t.sf(np.abs(t_stat),df)

##--------Using Build in Functions---------#

t_stat,p=stats.ttest_ind(group1,group2,equal_var=True) #equal_var=True means Equal variance

print(f"t-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"p-value: {p_value:.4f}")
```

#### Interpretation:

- Since p = 0.1543 > 0.05, we fail t0 reject the null hypothesis at the 5% significance level.
- There is no significant evidence that the group1 mean differs from the group2 mean.


#### Visualization

```python
from scipy.stats import t, norm

# Plot distributions
plt.figure(figsize=(12, 6))

# Plot normal distributions for both groups
x = np.linspace(min(min(group1), min(group2)) - 5, max(max(group1), max(group2)) + 5, 500)

# Plot group1
plt.hist(group1, bins=10, alpha=0.5, label='Group 1', color='skyblue', density=True)
# Plot group2
plt.hist(group2, bins=10, alpha=0.5, label='Group 2', color='salmon', density=True)

# Mean lines
plt.axvline(mean1, color='blue', linestyle='--', label=f'Mean 1: {mean1:.2f}')
plt.axvline(mean2, color='red', linestyle='--', label=f'Mean 2: {mean2:.2f}')

# Annotate t-statistic
plt.text((mean1+mean2)/2, 0.05, f"t = {t_stat:.2f}\np = {p_value:.4f}", 
         ha='center', bbox=dict(facecolor='white', edgecolor='black'))

plt.title('Independent Two-Sample t-Test: Distribution Comparison')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Paired Sample t-test

The paired sample t-test(also called dependent t-test) compares the mean of two related groups. 
This test is used when the same subjects are measured twice conditions, or at two different times.

```python

#creating two paired sample
np.random.seed(42)
#before: real value
before = data['BMXLEG'].sample(n=30, replace=True).reset_index(drop=True)
#after: same individuals, slightly changed
after = before + np.random.normal(loc=-0.5, scale=1.0, size=30)  
```

#### Hypothesis
- Null Hypothesis(H_0): mu_d = 0 (No change in mean)
- Alternative Hypothesis(H_1): mu_d != 0 (two tails)

  

```python
#-----------Manually------------------#
# difference 
diff= after - before

#mean & std of differences
d_bar= np.mean(diff)
std_d=np.std(diff,ddof=1)
n=len(diff)

# t-statistic
t_stat= d_bar / (std_d/ np.sqrt(n))

#df and p-value(two-tailed)
df=n-1
p_value= 2* stats.t.sf(np.abs(t_stat),df)

#----------Using Builtin Function------------#

t_stat, p_value = stats.ttest_rel(after,before)

print(f"Mean of differences: {d_bar:.4f}")
print(f"Standard deviation: {std_d:.4f}")
print(f"t-statistic: {t_stat:.4f}")
print(f"Degrees of freedom: {df}")
print(f"p-value: {p_value:.4f}")


```

#### Interpretation:
- Since p = 0.03 < 0.05, we  reject the null hypothesis at the 5% significance level.
- There is strong significant evidence that the difference in before and after.

```python
# Line plot of individual before-after values
plt.figure(figsize=(10, 6))
for i in range(len(before)):
    plt.plot(['Before', 'After'], [before[i], after[i]], marker='o', color='gray')
    
plt.title('Paired Sample - Before vs. After')
plt.ylabel('Measurement')
plt.grid(True)
plt.show()

# Histogram of differences
plt.figure(figsize=(8, 4))
plt.hist(diff, bins=5, color='purple', alpha=0.6, edgecolor='black')
plt.axvline(x=d_bar, color='red', linestyle='--', label=f'Mean diff = {d_bar:.2f}')
plt.title('Distribution of Differences (After - Before)')
plt.xlabel('Difference')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
